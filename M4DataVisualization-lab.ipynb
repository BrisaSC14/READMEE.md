## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY_TOP_DATABASES = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM master
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

top_databases_df = pd.read_sql_query(QUERY_TOP_DATABASES, conn)

plt.figure(figsize=(8, 8))
plt.pie(top_databases_df['count'], labels=top_databases_df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the corrected SQL query to calculate the median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY_MEDIAN = """
SELECT Age,
       ROUND((MAX(CASE WHEN Age BETWEEN 30 AND 35 THEN WorkWeekHrs END) + MIN(CASE WHEN Age BETWEEN 30 AND 35 THEN WorkWeekHrs END)) / 2, 2) AS Median_WorkWeekHrs,
       ROUND((MAX(CASE WHEN Age BETWEEN 30 AND 35 THEN CodeRevHrs END) + MIN(CASE WHEN Age BETWEEN 30 AND 35 THEN CodeRevHrs END)) / 2, 2) AS Median_CodeRevHrs
FROM master
GROUP BY Age
HAVING Age BETWEEN 30 AND 35
"""

# Execute the SQL query and read data into a DataFrame
median_df = pd.read_sql_query(QUERY_MEDIAN, conn)

# Close the database connection
conn.close()

# Display the resulting DataFrame
print(median_df)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the SQL query to calculate the median ConvertedComp for the age group 45 to 60
QUERY_MEDIAN = """
SELECT Age,
       ROUND((MAX(CASE WHEN Age BETWEEN 45 AND 60 THEN ConvertedComp END) + MIN(CASE WHEN Age BETWEEN 45 AND 60 THEN ConvertedComp END)) / 2, 2) AS Median_ConvertedComp
FROM master
GROUP BY Age
HAVING Age BETWEEN 45 AND 60
"""

# Execute the SQL query and read data into a DataFrame
median_df = pd.read_sql_query(QUERY_MEDIAN, conn)

# Plot the median ConvertedComp for all ages from 45 to 60
plt.figure(figsize=(10, 6))
plt.plot(median_df['Age'], median_df['Median_ConvertedComp'], marker='o', linestyle='-')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(median_df['Age'])
plt.show()

# Close the database connection
conn.close()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the SQL query to count the occurrences of each unique value in the MainBranch column
QUERY_MAIN_BRANCH_COUNTS = """
SELECT MainBranch, COUNT(*) AS Count
FROM master
GROUP BY MainBranch
"""

# Execute the SQL query and read data into a DataFrame
main_branch_counts_df = pd.read_sql_query(QUERY_MAIN_BRANCH_COUNTS, conn)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(main_branch_counts_df['MainBranch'], main_branch_counts_df['Count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of Respondents by MainBranch')
plt.grid(axis='x')
plt.show()

# Close the database connection
conn.close()
```
